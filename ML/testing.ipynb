```python
import pandas as pd

data = pd.read_csv('Datasets/sentiment140.csv', header=None, names=['polarity', 'id', 'date', 'query', 'user', 'text'])
data = data[['polarity', 'text']]
data
```

```python
import emoji

def contains_emoji(text):
    """Check if a text contains any emojis."""
    return any(char in emoji.EMOJI_DATA for char in text)
```

```python
def extract_emojis(text):
    """Extract all emojis from a given text."""
    return ''.join([char for char in text if char in emoji.EMOJI_DATA])
```

```python
# Check if each row contains an emoji
data['has_emoji'] = data['text'].apply(extract_emojis)

# Filter rows with emojis
rows_with_emojis = data[data['has_emoji'].str.len() > 0]

print(rows_with_emojis[['text', 'has_emoji']])
```
