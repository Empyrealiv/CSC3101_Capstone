```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import emoji
import torch
from gensim.models import keyedvectors
```

```python
data = pd.read_csv('Datasets/Cleaned/combined_dataset.csv')
data
```

```python
from collections import Counter

def count_emojis(text):
    return [char for char in text if char in emoji.EMOJI_DATA]

all_emojis = data['text'].apply(count_emojis).sum()

emoji_counts = Counter(all_emojis)

emoji_counts_df = pd.DataFrame(emoji_counts.items(), columns=['Emoji', 'Count'])
emoji_counts_df = emoji_counts_df.sort_values(by='Count', ascending=False).reset_index(drop=True)

print(emoji_counts_df)
```

```python
emoji_counts_df = emoji_counts_df[emoji_counts_df['Count'] >= 500]
emoji_list = emoji_counts_df['Emoji'].tolist()
emoji_list
```

```python
e2v = keyedvectors.load_word2vec_format('Embeddings/emoji2vec.bin', binary=True)
```

```python
emoji_vectors = {}
emoji_not_found_count = 0

for emoji in emoji_list:
    if emoji in e2v:
        emoji_vectors[emoji] = e2v[emoji]
    else:
        emoji_not_found_count += 1
        print(f"{emoji} not in vocabulary")

print(f"Number of emojis found in the vocabulary: {len(emoji_vectors)}")
print(f"Number of emojis not found in the vocabulary: {emoji_not_found_count}")
```

```python
for emoji, vector in emoji_vectors.items():
    print(f"{emoji}: {vector}")
```

```python
projection_layer = torch.nn.Linear(300, 768)
projected_emoji_vectors = {emoji: projection_layer(torch.tensor(vector).unsqueeze(0)).squeeze(0) for emoji, vector in emoji_vectors.items()} # Projecting the 300-dimensional vectors to 768 dimensions
```

```python
for emoji, vector in projected_emoji_vectors.items():
    print(f"Emoji: {emoji}, Projected Vector Shape: {vector.shape}")
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

emoji_presence = {}

for emoji in projected_emoji_vectors.keys():
    if emoji in tokenizer.vocab:
        emoji_presence[emoji] = True
    else:
        emoji_presence[emoji] = False

for emoji, present in emoji_presence.items():
    status = "present" if present else "not present"
    print(f"Emoji '{emoji}' is {status} in BERT's vocabulary.")
```

```python
# Test with words

word_list = ['happy', 'sad', 'angry', 'love', 'hate', 'good', 'bad', 'positive', 'negative', 'joy', 'fear', 'surprise', 'trust', 'disgust', 'anticipation']
word_presence = {}

for word in word_list:
    if word in tokenizer.vocab:
        word_presence[word] = True
    else:
        word_presence[word] = False

for word, present in word_presence.items():
    status = "present" if present else "not present"
    print(f"Word '{word}' is {status} in BERT's vocabulary.")
```

```python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

emojis_to_add = [emoji for emoji in projected_emoji_vectors.keys() if emoji not in tokenizer.vocab]

if emojis_to_add:
    tokenizer.add_tokens(emojis_to_add)
    model.resize_token_embeddings(len(tokenizer))

    embedding_layer = model.get_input_embeddings()
    
    for emoji in emojis_to_add:
        emoji_id = tokenizer.convert_tokens_to_ids(emoji)
        embedding_layer.weight.data[emoji_id] = projected_emoji_vectors[emoji]

    print("New emojis have been successfully added and initialized!")
else:
    print("All specified emojis are already in the tokenizer vocabulary.")
```

```python
# Verify that the embeddings have been added successfully

for emoji in emojis_to_add:
    emoji_id = tokenizer.convert_tokens_to_ids(emoji)
    emoji_embedding = model.get_input_embeddings().weight.data[emoji_id]
    print(f"Embedding for '{emoji}':", emoji_embedding[:5])
```

```python
# List of emojis that were added to the tokenizer vocabulary
emojis_to_check = ["üòÇ", "‚ù§Ô∏è", "üëç", "üî•", "ü•∫"]

# Check the token IDs for each emoji
for emoji in emojis_to_check:
    token_id = tokenizer.convert_tokens_to_ids(emoji)
    print(f"Emoji '{emoji}' has token ID: {token_id}")
```

```python
def load_glove_embeddings(file_path):
    embeddings_index = {}
    
    # Open the GloVe file and read line by line
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = vector
    
    print(f"Loaded {len(embeddings_index)} word vectors from GloVe.")
    return embeddings_index

file_path = 'Embeddings/glove.twitter.27B/glove.twitter.27B.200d.txt'
glove_embeddings = load_glove_embeddings(file_path)
```
